```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('Kuramoto_data.txt')

N = 10
num_oscillators = N * N
num_steps = data.shape[0]  

average_phase = np.zeros(num_steps)
order_parameter = np.zeros(num_steps, dtype=complex)

for t in range(num_steps):
    phases = data[t, :] 
    
    average_phase[t] = np.mean(phases)
    
    order_parameter[t] = np.mean(np.exp(1j * phases))

order_param_magnitude = np.abs(order_parameter)
order_param_phase = np.angle(order_parameter)

plt.figure(figsize=(14, 6))

# avg phase plotting
plt.subplot(1, 2, 1)
plt.plot(np.arange(num_steps), average_phase, label="Average Phase", color="blue")
plt.xlabel("Timestep")
plt.ylabel("Average Phase")
plt.title("Average Phase of Oscillators Over Time")
plt.legend()

# Order magnitude plotting
plt.subplot(1, 2, 2)
plt.plot(np.arange(num_steps), order_param_magnitude, label="Order Parameter Magnitude", color="red")
plt.xlabel("Timestep")
plt.ylabel("Order Parameter Magnitude")
plt.title("Order Parameter (Synchronization) Over Time")
plt.ylim(0, 1.2)
plt.legend()

plt.tight_layout()
plt.show()
```
