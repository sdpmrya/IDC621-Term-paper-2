```python
N = 10
tot_time = 1
dt = 0.0001
num_steps = int(tot_time/dt)

data.iloc[0]
```

```python
si = []
co = []
for i in range(0, num_steps):
    a = 0
    b = 0
    for j in range(0, N*N):
        a += mth.sin(data.iloc[i][j])
        b += mth.cos(data.iloc[i][j])
    if (i%100 == 0):
        si.append(a)
        co.append(b)
```

```python
plt.plot(si)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = np.loadtxt('Kuramoto_data.txt')

# Constants and Parameters
N = 10
num_oscillators = N * N
num_steps = data.shape[0]  # Each row in data represents one timestep

# Initialize arrays for calculations
average_phase = np.zeros(num_steps)
order_parameter = np.zeros(num_steps, dtype=complex)

# Calculate average phase and order parameter at each timestep
for t in range(num_steps):
    phases = data[t, :]  # Phases of all oscillators at timestep t
    
    # Average Phase (mean of phases across oscillators)
    average_phase[t] = np.mean(phases)
    
    # Order Parameter (magnitude and phase)
    order_parameter[t] = np.mean(np.exp(1j * phases))

# Convert order parameter to amplitude (magnitude) and phase (angle)
order_param_magnitude = np.abs(order_parameter)
order_param_phase = np.angle(order_parameter)

# Plotting the results
plt.figure(figsize=(14, 6))

# Plot average phase over time
plt.subplot(1, 2, 1)
plt.plot(np.arange(num_steps), average_phase, label="Average Phase", color="blue")
plt.xlabel("Timestep")
plt.ylabel("Average Phase")
plt.title("Average Phase of Oscillators Over Time")
plt.legend()

# Plot order parameter magnitude over time
plt.subplot(1, 2, 2)
plt.plot(np.arange(num_steps), order_param_magnitude, label="Order Parameter Magnitude", color="red")
plt.xlabel("Timestep")
plt.ylabel("Order Parameter Magnitude")
plt.title("Order Parameter (Synchronization) Over Time")
plt.legend()

plt.tight_layout()
plt.show()
```
